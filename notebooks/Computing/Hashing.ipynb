# Hashing & Hashing Algorithms

Counting is dependent on computation feasbility and physical motions like speed of light.
We have limited capacity of computation with speed of light being constant. Even if we do counting at light speed. (unless we make use of dark speed)

Hashing can account for available limited computation capacity.
Hashing is the only way to count only items which is of our interest or decidability instead of everything.


Hashing mostly makes use of "Pigeon Hole Principles" & "Extended Pigeon Hole Principle" for categorization.

Hashing is highly dependent on NUMBER THEORY.

Modulo Arithmetic is a kind of Hashing function which categorizes based on Reminder.

Concepts of Hashing:

1. Hash Function
2. Hash Table
3. Hash Data Distribution or Locality
4. Collision
5. Collision Resolution
6. Probing (linear & quadratic)


Special types of Hashing
1. Bloom Filter (Probailistic data structure)
2. Geometric
3. Cryptographic

Algorithms
1. Rabin Karp
2. Knuth–Morris–Pratt Algorithm
https://brilliant.org/wiki/rabin-karp-algorithm/

Common Hashing
1. Modulo
2. Truncation or Digit/Character Extraction (like dictionary or library book arrangment)
3. Splitting Keys
4. Radix Conversion (like log, hex, octa)
5. Mid Square hashing (square a number and take middle part)
6. Random Number generator

Common Applications
1. Database systems
2. Symbol tables
3. Data Dictionaries
4. Network Processing Algo (route lookup, packet classification, n/w monitoring)
5. Browser Caches

Associative arrays: Hash tables are commonly used to implement many types of in-memory tables. They are used to implement associative arrays (arrays whose indices are arbitrary strings or other complicated objects).

Database indexing: Hash tables may also be used as disk-based data structures and database indices (such as in dbm).

Caches: Hash tables can be used to implement caches i.e. auxiliary data tables that are used to speed up the access to data, which is 
primarily stored in slower media.

Object representation: Several dynamic languages, such as Perl, Python, JavaScript, and Ruby use hash tables to implement objects.
Hash Functions are used in various algorithms to make their computing faster

Not Suited places
1. Sorting where ORDINAL ordering is required.
2. Multidimensional data
3. Prefix searching
4. dynamic data and no unique keys


Ways of generation of Hash
a. Static
b. Dynamic (size of table can grow, hash function should handle it)
Load factor is important measure over here

Hash Table Types (or called as scatter table)
1. Open Addressed Hash Tables (Probing - Linear & Quardratic, )
2. Chained Hash Table


Important key aspect which we cannot ignore on Hash Function.
1. It should not consume much time and space.
2. Pigeon hole principle
3. Data distribution can be well represented by Hash distribution (based on ordinality) - Conversely if we know a distribution of data we can write a better hash function.
